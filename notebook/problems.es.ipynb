## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random
import numpy as np

Total_lanzamientos = 1000

contador = 0

# Para simular cada lanzamiento como son un numero conocido
# es adecuado usar un for
for i in range(Total_lanzamientos): 

    # Simulacion del lanzamiento del dado 1
    dado1 = random.randint(1, 6)

    # Simulacion del lanzamiento del dado 2
    dado2 = random.randint(1, 6)
        
    # Suma de los resultados del lanzamiento para definir el primer evento de interes
    suma = dado1 + dado2
        
    # Condicional que permite acumular en el contador un evento positivo
    # El el evento suma par o suma mayor a 7
    if (suma % 2 == 0) or (suma > 7):
        contador = contador + 1

# Resultado de la probabilidad

Probabilidad = contador / Total_lanzamientos

print(f"La probabilidad de obtener un número par o mayor que 7 es: {np.round(Probabilidad*100, 2)}%")
```
